```python
import pandas as pd

# Load the dataset
df = pd.read_csv("flyrank-ml-internship/data/raw/content_refresh_anonymized.csv")

# Basic information
print("Dataset Shape:", df.shape)
print("Number of Rows:", len(df))
print("Number of Columns:", len(df.columns))

# Statistics
print("\nAverage Impressions (90d):", round(df["impressions_90d"].mean(), 2))
print("Average Clicks (90d):", round(df["clicks_90d"].mean(), 2))
print("Average Sessions (90d):", round(df["sessions_90d"].mean(), 2))

# Trend distribution
print("\nTrend Direction Counts:")
print(df["trend_direction"].value_counts())
```
